```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Roots (x+0.02h)(x+(400+4.42h))

def b(h):
    return 100+1.11*h

def a(h):
    return (1/4)

def c(h):
#     return (10000+1.11*h)**2-(10000+1.1*h)**2
    return 0.01*h*(200+2.21*h)

# Given numpy array h calculates the smaller root which involves cancellation error.
def cal(h):
    return ((-b(h)+np.sqrt(b(h)**2-4*a(h)*c(h)))/(2*a(h)))

def cal2(h):
    return ((2*c(h))/(-b(h)-np.sqrt(b(h)**2-4*a(h)*c(h))))

def xStar(h):
    return -0.02*h
```

```python
h=np.linspace(1e-2,1e-5,1e4)
x=h
y1=(abs(xStar(h)-cal(h)))
y2=(abs(xStar(h)-cal2(h)))
```

```python
xlog = (np.log(x[:100]))
ylog1 = (np.log(y1[:100]))
ylog2 = (np.log(y2[:100]))
ax = plt.gca()
plt.plot(xlog, ylog1, linewidth=2, color='navy')
plt.legend(loc='best')
plt.xlabel(r'log(h)')
plt.ylabel(r'log(y)')
ax.grid(True)
plt.title(r'Log-Log plot')
plt.show()
plt.clf()
```
